```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Apartment prices

read in data

```python
df = pd.read_csv('https://raw.githubusercontent.com/su-mt4007/data/refs/heads/main/Booli_sold.csv')
df.head()
```

## calculating ppsqm for each apartment sold

```python
df['ppsqm'] = df['soldPrice'] / df['livingArea']
df.head()
```

## top 5 expensive appartments

```python
top_5 = df.sort_values(by='ppsqm', ascending=False).head(5)
top_5
```

## Ekhagen avg square price

```python
average_ppsqm = df['ppsqm'].mean()
print(average_ppsqm)
```

## Interesting data

```python
df = df[df['livingArea'] > 0]
df = df[df['constructionYear'].notna()]

# Create a new column for properties built before and after 1950
df['pre_1950'] = df['constructionYear'] < 1950

# Calculate the average price per square meter for both categories
average_ppsqm_by_year = df.groupby('pre_1950')['ppsqm'].mean()

# Visualize the comparison
plt.figure(figsize=(8, 5))
average_ppsqm_by_year.plot(kind='bar', color=['orange', 'blue'])
plt.title('Average ppsqm: homes built before vs after 1950', fontsize=14)
plt.ylabel('ppsqm (SEK)', fontsize=12)
plt.xticks([0, 1], ['Before 1950', '1950 and after'], rotation=0, fontsize=11)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

intresting to look whether older homes built before 1950 sell for more money because of unique designs or history

# Swedish Election of 2018

## Read data

```python
df_election = pd.read_csv("https://raw.githubusercontent.com/su-mt4007/data/refs/heads/main/2018_R_per_kommun.csv", delimiter=';', decimal=',')
df_election.head()
```

## Calculating total number of legitimate votes

```python
stockholm_data = df_election[df_election['LÄNSNAMN'] == 'Stockholms län']
total_votes = stockholm_data['RÖSTER GILTIGA'].sum()
print(f"total number of legitimate votes in Stockholm: {total_votes}")
```

## Municipality with highest voting percentage for S

```python
s = df_election.sort_values(by='S', ascending=False)
highest = s.head(1)
print(highest['KOMMUNNAMN'].at[highest.index[0]])
```

## Top 3 highest municipalities participation

```python
top_3 = df_election.sort_values(by='VALDELTAGANDE', ascending=False).head(3)
print(top_3[['KOMMUNNAMN', 'VALDELTAGANDE']])
```

# Predicting strokes

```python
stroke_data = pd.read_csv("https://raw.githubusercontent.com/su-mt4007/data/refs/heads/main/stroke-data.csv")

factor_combinations = stroke_data.groupby(['hypertension', 'smoking_status', 'Residence_type'])['stroke'].mean().unstack()

factor_combinations.plot(kind='bar', figsize=(12, 7), color=['gold', 'blue'], edgecolor='black')
plt.title('Stroke proportion based on hypertension, smoking status, and residence', fontsize=14)
plt.xlabel('Hypertension and smoking status', fontsize=12)
plt.ylabel('Stroke proportion', fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.legend(title='Residence Type', fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

The graph indicates that hypertension and smoking are the strongest contributors to stroke risk and rural/urban having less noticeable effect, but hypertension in combination with smoking seems to create the highest risk.
