### Research Questions

The focus of this project is on responding to three questions:

- Are there significant differences in the ratings given by video game users versus critics?
- How do critic and user ratings influence the commercial success of a video game, either boosting or hindering its sales?
- How do preferences for different genres vary across regions?

### Dataset

For this analysis, I used the dataset 'Video Game Sales and Ratings' created by Sumit Kumar Shukla, available on [Kaggle.com](https://www.kaggle.com/datasets/thedevastator/video-game-sales-and-ratings), used under the Open Data Commons Attribution (ODC-BY) license, which provided a comprehensive look at video game sales spanning from 1985 to 2016. The dataset included detailed information for each game, such as its title, platform, genre, release year, global sales figures, and both critic and user ratings.

To dive deeper into the analysis, I created three additional features designed to shed light on key research questions. These included a measure of the score disparity between critics and users, along with categorized versions of both critic and user ratings to better understand trends. Once the data was imported and these new features were generated, the next step was cleaning. This involved merging the datasets based on game titles, removing missing or null values, and eliminating any duplicate entries. After cleaning and organizing the data, the final dataset contained a little over 7,000 entries and 17 columns of useful information, ready for further exploration.


### Data Analysis

To start with my analysis, I began with creating a visualization of global video game sales categorized by genre. The plot revealed that Action, Sports and Shooters games were the leading genres in sales, while Adventure, Puzzle, and Strategy games ranked the lowest.

I then examined how crtics' scores for video games compared to those given by users. I created a box plot that displayed the median critic score for each genre, and sorted in descending order to visualize this. The box plot shows that critics gave the highest ratings to Sports, Strategy, and Role-Playing games, whereas the genres receiving the lowest scores were Adventure, Action, and Platform.

For User Scores, again using the box plot shows that Role-Playing, Strategy, and Platform got the highest ratings while Misc, Shooter, and Action recieved the lowest. 

To explore the disparity between users' and critics' ratings, a new column was created to show the score differences. This was done by subtracting the critic score from the user score for each video game title, then grouping the results by genre. The analysis revealed that, on average, users rated Adventure, Platform, and Fighting games higher than critics, which instead showed a stronger preference for Sports games.

With the analysis of video game scores from both users and critics completed, we turn to the second question: How do critic and user ratings influence the commercial success of a video game, either boosting or hindering its sales? Generally, we can observe that there is a positive correlation between the scores and sales. The higher the scores are, the higher the global sales tend to be. The Critic Score shows a stronger correlation with global sales than the User Score, as evidenced by the steeper slope of the Critic Score line. This suggests that increases in critic scores have a greater impact on sales than user scores. Additionally, even at lower scores, sales for the Critic Score are lower than for the User Score, indicating that critic ratings have a stronger negative influence on sales at lower levels.

Lastly, how do preferences for different genres vary across regions? Given that North America and Europe have larger populations than Japan, it was expected that the regional proportions would differ. 

The chart highlights distinct preferences for game genres across regions, suggesting that cultural tastes play a more significant role than population differences alone. North America shows a strong preference for genres like Shooter and Sports, indicating a cultural inclination towards action-packed and competitive games. Europe follows a similar trend but with slightly less emphasis on Shooters and a relatively higher share in Racing games. Japan stands out with a markedly higher proportion of sales in Role-Playing and Puzzle genres, reflecting a cultural preference for games with strategic depth and rich narratives, contrasting sharply with the action-oriented focus in Western markets.

Given these variations, a targeted regional approach to marketing would be highly effective. Certain games may achieve immense success in one region while struggling to gain traction elsewhere, depending on their genre and appeal. By tailoring marketing strategies and product offerings to align with regional tastes, developers and publishers can maximize the potential for success in each market.

### Conclusion

With growing interest driving market expansion and new technologies, the video game industry has evolved significantly over the past decade. Despite this growth, it seems the market isn't fully aligning with the preferences of users. While ratings play a significant role in driving sales, it becomes evident that critic scores tend to have a more substantial impact on a game’s commercial success than user scores. Games with higher critic ratings generally see greater global sales, with the effect being more pronounced at higher and lower score extremes.
